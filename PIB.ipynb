```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h, m = 1, 1 
hbar = h/(2*np.pi)

def Epib1d(n=1, L=1):
        '''Calculates energy of 1D PIB state.
        Args:
        n (int): Quantum number specifying state.
        L (float): Box dimensions in Bohr.
        Returns:
         Float energy value in Ha.
        '''

        En = (n**2 * np.pi**2 * hbar**2) / (2 * m * L**2)

        return En

    
def psi(x, n=1, L=1):      
        '''Numeric representation of the normalized 1D PIB wavefunction.
        Args:
         x  (float or array): Cartesian spatial values.
         n  (int):   Quantum number
         L  (float): Box dimensions in Bohr.
        Returns:
        Wavefunction evaulated at a point or array if input x are arrays.
        '''

        fn = np.sqrt(2/L)*np.sin(n*np.pi*x/L)

        return fn
```

```python
### Change these values
n=4
L=1

### Compute wavefunction and its square
x = np.linspace(0, L, 100)
psi1d = psi(x, n=n, L=L)

### Plot wave function and its square
plt.plot(x, psi1d, lw=2, label=r'$\psi_{n}(x)$')
plt.plot(x, psi1d**2, lw=2, label=r'$\psi_{n}(x)^2$')
plt.grid('--')
plt.legend()
plt.title(f'n={n}')
```

```python
xx, yy = np.meshgrid( np.linspace(-1,1,100),np.linspace(-1,1,100) )
```

```python
fig, ax = plt.subplots(subplot_kw={'projection':'3d'})
ax.plot_surface(xx, yy, psi(xx,2)*psi(yy), cmap='viridis')
ax.view_init(60, 30)

# Adjust the padding for the ticks on x and y axis
ax.tick_params(axis='x', pad=-2)  # Adjust this value to bring xticks closer
ax.tick_params(axis='y', pad=-2.5)  # Adjust this value to bring yticks closer

```
