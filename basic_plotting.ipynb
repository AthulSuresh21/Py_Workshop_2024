```python
def f(t):
    return np.exp(-t**2) * np.cos(5*np.pi*t)

t1 = np.arange(-5., 5., .1)
t2 = np.arange(-5., 5., .02)

plt.figure(1, figsize=(12,12))
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
plt.title('Gaussian wavepacket 1')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'v--')
plt.title('cos(x)')
plt.show()

```

```python
def log_map(x,a):
    return a-x*x

num_a = 10000
num_discard = 300
num_keep = 1000

a = np.linspace(-0.25, 2, num_a)
x = np.ones_like(a)*0.1

#Call figure to be saved onto
fig= plt.figure(figsize=(12,12))

#Number of transients
for i in range(num_discard):
    x = log_map(x,a)

for i in range(num_keep):
    plt.plot(a,x,'k,',alpha=0.01)
    x = log_map(x,a)

plt.title("Bifurcation Diagram for Logistic Map")
plt.xlabel('$a$',fontsize=15)
plt.ylabel('$x$',fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
#fig.savefig('log.eps')
plt.show()

```

```python
x = np.linspace(-10, 10, 1000)
y = np.linspace(-10, 10, 1000)

xv, yv = np.meshgrid(x, y)
```

```python
f_xy = (xv**2 + yv**2)*np.sin(3*xv)*np.cos(3*yv)
```

```python
plt.contourf(xv, yv, f_xy,levels=50)
plt.colorbar()
plt.show()

```

```python
f = np.exp(-(xv**2+yv**2)/6) * np.sin(xv)
```

```python
fig = plt.figure(figsize = (8,8))
ax = plt.axes(projection='3d')
ax.grid()

ax.plot_surface(xv, yv, f,cmap='OrRd')
ax.set_title('$e^{-(x^2 + y^2)/6}sin(x)$')

# Set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('t', labelpad=20)

plt.show()
```
